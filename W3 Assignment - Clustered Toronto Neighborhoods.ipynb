Retrieve HTML page contents using BeautifulSoup4

```python
URL = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"

res = requests.get(URL).text
soup = BeautifulSoup(res,'lxml')
print(soup.prettify())
```

Retrieve only the relevant webpage section (i.e. inside the table HTML tag)

```python
postcode_table = soup.find('table',{'class':'wikitable'})
td_raw = postcode_table.findAll('td')
td_raw
```

Convert retrieved data into a list of strings
1st cleanup step: split the list along the HTML table delimiters

```python
td_raw3 = str(td_raw)
td_raw3 = td_raw3.split("\n</td>, <td>")
td_raw3
```

Interval-slicing the postcodes, boroughs, and neighborhoods into their own lists

```python
td_raw3_postcodes = td_raw3[::3] #extract postcodes into a single-dimension list
td_raw3_postcodes[0] = td_raw3_postcodes[0][-3 : len(td_raw3_postcodes[0])] #remove HTML tag on 0th list element
td_raw3_postcodes
```

```python
td_raw3_boroughs = td_raw3[1::3]
td_raw3_boroughs
```

```python
td_raw3_neigh = td_raw3[2::3]
td_raw3_neigh
```

Cleaning stray HTML characters from the final list element

```python
last_elem = len(td_raw3_neigh) - 1
td_raw3_neigh[last_elem] = td_raw3_neigh[last_elem].strip("\n")
td_raw3_neigh[last_elem] = td_raw3_neigh[last_elem].strip("</td>")
td_raw3_neigh[last_elem] = td_raw3_neigh[last_elem].strip("]")
td_raw3_neigh[last_elem]
```

Replacing forward-slash separators (from Wikipedia table) with commas (Assignment specification)

```python
td_raw3_neigh = [c.replace(' /', ',') for c in td_raw3_neigh]
td_raw3_neigh
```

Preparing to combine the lists into a dataframe, via an intermediate dictionary

```python
dict_fsa = {'PostalCode': td_raw3_postcodes, 'Borough': td_raw3_boroughs, 'Neighborhood': td_raw3_neigh}
dict_fsa
```

Converting into a pandas df

```python
import pandas as pd
df_fsa = pd.DataFrame(dict_fsa)
df_fsa
```

Removing postcodes without assigned boroughs, per Assignment specifications

```python
df_fsa_assigned = df_fsa[df_fsa['Borough'] != 'Not assigned']
df_fsa_assigned.reset_index(drop=True, inplace=True)
df_fsa_assigned
```

Quick check if any data points match the criteria of having a Borough but not having a Neighborhood. Per the empty df below, looks like there aren't any.

```python
df_blanksearch = df_fsa_assigned[df_fsa_assigned['Neighborhood'] == '']
df_blanksearch
```

In case data will change in the future, adding a forward-fill for postcodes with assigned boroughs but without assigned neighborhoods.

```python
df_final = df_fsa_assigned.ffill(axis = 1)
```

```python
df_final.shape
```

```python
df_coords = pd.read_csv('http://cocl.us/Geospatial_data')
```

```python
df_coords.rename(columns={'Postal Code':'PostalCode'}, inplace=True)
df_coords.head()
```

```python
df_postal_coords = df_final.merge(df_coords, on='PostalCode')
df_postal_coords
```

Importing clustering package

```python
from sklearn.cluster import KMeans
```

Defining a variable with only the values for the clustering

```python
X = df_postal_coords.iloc[0:len(df_postal_coords.index)+1,3:5]
X
```

Implementing the clustering; I'd left this open ended, so that I could change the k value as I wanted

```python
k = 3
k_means = KMeans(init = "k-means++", n_clusters = k, n_init = 12)

k_means.fit(X)
labels = k_means.labels_
print(labels)
```

Adding the cluster labels back to the original reference df

```python
df_clustered = df_postal_coords
df_clustered['Cluster'] = labels
df_clustered.head(5)
```

Converting the latitude, longitude, and cluster ID values into a list usable by Folium.
I realize in hindsight I could have done this with X (the modeling variable) instead, but oh well.

```python
df_clus0 = df_clustered[df_clustered['Cluster']==0]
df_clus0 = df_clus0[['Latitude','Longitude']]
list_clus0 = df_clus0.values.tolist()

df_clus1 = df_clustered[df_clustered['Cluster']==1]
df_clus1 = df_clus1[['Latitude','Longitude']]
list_clus1 = df_clus1.values.tolist()

df_clus2 = df_clustered[df_clustered['Cluster']==2]
df_clus2 = df_clus2[['Latitude','Longitude']]
list_clus2 = df_clus2.values.tolist()

if k >=4:
    df_clus3 = df_clustered[df_clustered['Cluster']==3]
    df_clus3 = df_clus3[['Latitude','Longitude']]
    list_clus3 = df_clus3.values.tolist()
```

Importing needed packages for mapping

```python
import os
import folium
print(folium.__version__)
```

Applying the clustered coordinates as points on the map.

Ended up making just 3 clusters instead of 4, as there wasn't (visually) a big improvement in mean distances between points (vs k=4).

There might be a more significant improvement with higher k (7, maybe?), but there really isn't any objective function to optimize in the Assignment.

```python
from folium.plugins import MarkerCluster

view_lat = df_clustered['Latitude'].median()
view_long = df_clustered['Longitude'].median()

map = folium.Map(location=[view_lat, view_long], tiles='CartoDB positron', zoom_start=11)

for point in range(0,len(list_clus0)):
    folium.Marker(list_clus0[point], folium.Icon(icon_color='white')).add_to(map)

for point in range(0,len(list_clus1)):
    folium.Marker(list_clus1[point], folium.Icon(icon_color='blue')).add_to(map)

for point in range(0,len(list_clus2)):
    folium.Marker(list_clus2[point], folium.Icon(icon_color='yellow')).add_to(map)    

if k >= 4:
    for point in range(0,len(list_clus3)):
        folium.Marker(list_clus3[point], folium.Icon(icon_color='green')).add_to(map)

map
```
