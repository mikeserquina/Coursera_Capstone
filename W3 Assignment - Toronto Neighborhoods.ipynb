Retrieve HTML page contents using BeautifulSoup4

```python
URL = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"

res = requests.get(URL).text
soup = BeautifulSoup(res,'lxml')
print(soup.prettify())
```

Retrieve only the relevant webpage section (i.e. inside the table HTML tag)

```python
postcode_table = soup.find('table',{'class':'wikitable'})
td_raw = postcode_table.findAll('td')
td_raw
```

Convert retrieved data into a list of strings
1st cleanup step: split the list along the HTML table delimiters

```python
td_raw3 = str(td_raw)
td_raw3 = td_raw3.split("\n</td>, <td>")
td_raw3
```

Interval-slicing the postcodes, boroughs, and neighborhoods into their own lists

```python
td_raw3_postcodes = td_raw3[::3] #extract postcodes into a single-dimension list
td_raw3_postcodes[0] = td_raw3_postcodes[0][-3 : len(td_raw3_postcodes[0])] #remove HTML tag on 0th list element
td_raw3_postcodes
```

```python
td_raw3_boroughs = td_raw3[1::3]
td_raw3_boroughs
```

```python
td_raw3_neigh = td_raw3[2::3]
td_raw3_neigh
```

Cleaning stray HTML characters from the final list element

```python
last_elem = len(td_raw3_neigh) - 1
td_raw3_neigh[last_elem] = td_raw3_neigh[last_elem].strip("\n")
td_raw3_neigh[last_elem] = td_raw3_neigh[last_elem].strip("</td>")
td_raw3_neigh[last_elem] = td_raw3_neigh[last_elem].strip("]")
td_raw3_neigh[last_elem]
```

Replacing forward-slash separators (from Wikipedia table) with commas (Assignment specification)

```python
td_raw3_neigh = [c.replace(' /', ',') for c in td_raw3_neigh]
td_raw3_neigh
```

Preparing to combine the lists into a dataframe, via an intermediate dictionary

```python
dict_fsa = {'PostalCode': td_raw3_postcodes, 'Borough': td_raw3_boroughs, 'Neighborhood': td_raw3_neigh}
dict_fsa
```

Converting into a pandas df

```python
import pandas as pd
df_fsa = pd.DataFrame(dict_fsa)
df_fsa
```

Removing postcodes without assigned boroughs, per Assignment specifications

```python
df_fsa_assigned = df_fsa[df_fsa['Borough'] != 'Not assigned']
df_fsa_assigned.reset_index(drop=True, inplace=True)
df_fsa_assigned
```

Quick check if any data points match the criteria of having a Borough but not having a Neighborhood. Per the empty df below, looks like there aren't any.

```python
df_blanksearch = df_fsa_assigned[df_fsa_assigned['Neighborhood'] == '']
df_blanksearch
```

In case data will change in the future, adding a forward-fill for postcodes with assigned boroughs but without assigned neighborhoods.

```python
df_final = df_fsa_assigned.ffill(axis = 1)
```
